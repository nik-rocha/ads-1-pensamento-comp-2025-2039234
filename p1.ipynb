1) Escreva um programa que leia um número e verifique se é ou não um número primo. (0,5 pontos)


```python
num = int(input("Digite um número: "))

if num <= 1:
    print("O número não é primo!")
else:
    prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            prime = False
            break
    
    if prime:
        print("O número é primo!")
    else:
        print("O número não é primo!")
```

O programa deverá:

● Exibir “Código inválido” caso o usuário digite um código não existente na tabela.

```python
purchaseValue = 0

while True:
    productCode = int(input('Digite o código do produto de 1-5 e digite 0 para ver o total. Compre o quanto quiser!: '))

    match productCode:
        case 1:
            productValue = 0.50
            productQuant = int(input('Digite a quantidade desse produto: '))
            if productQuant <= 0:
                print('Quantidade inválida!')
            purchaseValue += productValue * productQuant
            print(f'Valor atual: R${purchaseValue:.2f}')
        case 2:
            productValue = 1.00
            productQuant = int(input('Digite a quantidade desse produto: '))
            if productQuant <= 0:
                print('Quantidade inválida!')
            purchaseValue += productValue * productQuant
            print(f'Valor atual: R${purchaseValue:.2f}')
        case 3:
            productValue = 4.00
            productQuant = int(input('Digite a quantidade desse produto: '))
            if productQuant <= 0:
                print('Quantidade inválida!')
            purchaseValue += productValue * productQuant
            print(f'Valor atual: R${purchaseValue:.2f}')
        case 4:
            productValue = 7.00
            productQuant = int(input('Digite a quantidade desse produto: '))
            if productQuant <= 0:
                print('Quantidade inválida!')
            purchaseValue += productValue * productQuant
            print(f'Valor atual: R${purchaseValue:.2f}')
        case 5:
            productValue = 8.00
            productQuant = int(input('Digite a quantidade desse produto: '))
            if productQuant <= 0:
                print('Quantidade inválida!')
            purchaseValue += productValue * productQuant
            print(f'Valor atual: R${purchaseValue:.2f}')
        case 0:
            print(f'O valor da compra ficou R${purchaseValue:.2f}, obrigado pela compra!')
            break
        case _:
            print(f'Código inválido: {productCode}')
```

    ○ Derrota (+0 pontos)

● Mostrar o total de pontos ao fim de cada rodada.

● Permitir que o usuário encerre o programa digitando 0.

● Ao finalizar, exiba a pontuação total do time no campeonato mostrando quantas vitórias, empates e derrotas o time teve e sua classificação ou rebaixamento:

    ○ Libertadores (>20 pontos)

    ○ Sulamericana (> 15 pontos)

    ○ Rebaixamento (< 5 pontos)

```python
teamName = input('Digite o nome do time do Brasileirão')
teamPoints = 0
x = 0
victories = 0
draws = 0
defeats = 0

while True:
    x += 1
    teamResult = int(input(f'Rodada {x}: Qual foi o resultado da partida?\n1- Vitória\n2- Empate\n3- Derrota\nDigite 0 para encerrar.'))
    
    if teamResult == 0:
        if teamPoints > 20:
            print(f'Seu time foi para a Libertadores com {teamPoints} pontos e {victories} vitórias, {draws} empates e {defeats} derrotas!')
            break
        elif teamPoints > 15:
            print(f'Seu time foi para a Sulamericana com {teamPoints} pontos e {victories} vitórias, {draws} empates e {defeats} derrotas!')
            break
        elif teamPoints < 5:
            print(f'Seu time foi para o rebaixamento com {teamPoints} pontos e {victories} vitórias, {draws} empates e {defeats} derrotas!')
            break
        else:
            print('Pontos inválidos.')
            break

    elif teamResult == 1:
        teamPoints += 3
        victories += 1
        print(f'Total de pontos atuais: {teamPoints}')

    elif teamResult == 2:
        teamPoints += 1
        draws += 1
        print(f'Total de pontos atuais: {teamPoints}')

    elif teamResult == 3:
        defeats += 1
        print(f'Nenhum ponto adicionado. Total de pontos atuais: {teamPoints}')

    else:
        print('Resultado inválido.')
```
